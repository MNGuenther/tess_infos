## Searching for specifics

`cat.get()` returns a pandas DataFrame subcatalog based on specific information you request.

All available arguments to add: `tic_id` (int), `sector` (int or list), `keywords` (str or list).

```python
## For only GAIA DR2 information
cat.get(tic_id=2733208, sector=2, keywords=['GAIADR2'])
```

```python
## For only TIC v8 information
cat.get(tic_id=2733208, sector=2, keywords=['TICv8'])
```

```python
## For only BANYAN-SIGMA information
cat.get(tic_id=2733208, sector=2, keywords=['BANYAN'])
```

```python
## For all targets in sectors 4 and 5
cat.get(sector=[4,5])
```

```python
## For more specific stellar information
cat.get(sector=4, keywords='radial_velocity')
```
